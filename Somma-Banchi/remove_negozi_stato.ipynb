```python
# Import Required Libraries
import json
from pathlib import Path
import re
```

```python
# Read Raw JSON File
path = Path('Somma-Banchi/data.json')
raw_text = path.read_text(encoding='utf-8')
print('Loaded raw text length:', len(raw_text))
```

```python
# Preprocess Invalid JSON
clean_text = re.sub(r',\s*([}\]])', r'\1', raw_text)
# Remove inline comments if any
clean_text = re.sub(r'//.*', '', clean_text)
clean_text = re.sub(r'/\*.*?\*/', '', clean_text, flags=re.S)
print('Cleaned text length:', len(clean_text))
```

```python
# Parse JSON into Python Objects
data = json.loads(clean_text)
print('Parsed items count:', len(data.get('items', [])))
```

```python
# Remove `Negozi-stato` from Each Item
for item in data.get('items', []):
    item.pop('Negozi-stato', None)
print('Removed Negozi-stato from items.')
```

```python
# Write Cleaned JSON Back to File
with path.open('w', encoding='utf-8') as f:
    json.dump(data, f, ensure_ascii=False, indent=2)
print('Wrote cleaned JSON to', path)
```

```python
# Verify Removal
found = [item for item in data.get('items', []) if 'Negozi-stato' in item]
print('Items still containing Negozi-stato:', len(found))
if found:
    print(found[:5])
```
